```python
import pandas as pd

df = pd.read_csv('C:/Users/djenb/Desktop/Portfolio/Python_health_data_project/data/raw/heart_rate_tracker.csv', header=1)

# Inspect the data
df.shape
df.columns
df.head()
df.dtypes

df.columns = df.columns.str.replace('com.samsung.shealth.tracker.heart_rate.', '')
df.columns = df.columns.str.replace('com.samsung.health.heart_rate.', '')

# Columns to keep
df = df[['start_time', 'heart_rate', 'time_offset']]

# Convert start_time to datetime format
df['start_time'] = pd.to_datetime(df['start_time'])

# Rename columns for readability
df = df.rename(columns={'start_time': 'recorded_at', 'heart_rate': 'BPM', 'time_offset': 'timezone'})

# Check for missing values
df.isna().sum()

# Check for suspicious BPM readings
len(df[(df['BPM'] >= 200) | (df['BPM'] <= 40)])

df.to_csv('C:/Users/djenb/Desktop/Portfolio/Python_health_data_project/data/cleaned/heart_rate_clean.csv', index=False)

```

## Heart Rate Data Cleaning
### Source: Samsung Health Export
### Goal: Clean raw heart rate data for comparison against public health benchmarks

## Step 1: Import Libraries
Imported pandas for data manipulation.

## Step 2: Load Data
Loaded raw heart rate CSV exported from Samsung Health.
header=1 was used as Samsung added an extra row above the real column headers.

## Step 3: Data Inspection
Raw data contains 3000+ rows from February 2024 to March 2026.
Column names contained Samsung's internal prefix which was removed in Step 4.

## Step 4: Column Cleaning
Removed com.samsung.shealth.tracker.heart_rate. and com.samsung.health.heart_rate. prefixes from all column names.

## Step 5: Column Selection
Kept only informative columns start_time, heart_rate and time_offset.
min_heart_rate and max_heart_rate were dropped as 99% of values were missing.

## Step 6: Convert DateTime
Converted recorded_at from plain text to a proper datetime format so Python can sort and filter by date.

## Step 7: Column Renaming
Renamed columns for readability and consistency across all notebooks.
start_time → recorded_at
heart_rate → BPM
time_offset → timezone

## Step 8: Data Validation
No missing values found in any column.
No suspicious BPM readings found outside the 40-220 range.
Data is clean and ready for comparison.

## Step 9: Save Cleaned Data
Saved cleaned data to data/cleaned/heart_rate_clean.csv.
