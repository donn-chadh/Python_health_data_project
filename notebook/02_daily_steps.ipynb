```python
import pandas as pd 

df = pd.read_csv('C:/Users/Donnchadh/Desktop/Portfolio/Python_health_data_project/data/raw/daily_steps.csv', header=1)

# Inspect the data
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.head(1))
print(df['calorie'].describe())
print(df['speed'].describe())
print(df['count'].describe())
print(df['distance'].describe())
print(df.isna().sum())

# Columns to keep + rename
df = df[['create_time', 'count', 'speed', 'distance', 'calorie']]
df = df.rename(columns={'create_time': 'start_time', 'count': 'steps', 'distance': 'distance_m', 'calorie': 'calories', 'speed': 'speed_ms'})

# Convert datetime and extract date and time
df['date'] = pd.to_datetime(df['start_time']).dt.date
df['start_time'] = pd.to_datetime(df['start_time']).dt.time

# Rounding columns for readability
df['distance_km'] = round(df['distance_m'] / 1000, 2)
df = df.drop(columns=['distance_m', 'speed_ms'])
df['calories'] = round(df['calories'], 1)

print(f"Validating rounding:\n{df[['start_time', 'date', 'distance_km', 'calories']].head(1)}")

# Save to CSV file
df.to_csv('../data/cleaned/daily_steps_clean.csv', index=False)
```
