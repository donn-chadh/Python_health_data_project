```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("arashnic/fitbit")
print("Path to dataset files:", path)

# Read in CSV to DF
df = pd.read_csv('C:/Users/Donnchadh/Desktop/Portfolio/Python_health_data_project/data/raw/fitbit_daily_activity.csv')

# Inspect the data
print(df.shape)
print(df.columns)
print(df.head())
print(df.dtypes)
print(df.isna().sum())

# Select columns to keep
df = df[['ActivityDate', 'TotalSteps', 'TotalDistance', 'Calories']]

# Rename columns with fitbit abbreviation
df = df.rename(columns={'ActivityDate': 'fb_date', 'TotalSteps': 'fb_total_steps', 'TotalDistance': 'fb_total_distance_km', 'Calories': 'fb_calories'})

# Validate name changes
print(df.columns)

# Check for obscure data ranges
print(df['fb_total_steps'].describe())
print(df['fb_total_distance_km'].describe())
print(df['fb_calories'].describe())

# Save to CSV file
df.to_csv('../data/cleaned/kaggle_steps.csv', index=False)
```
