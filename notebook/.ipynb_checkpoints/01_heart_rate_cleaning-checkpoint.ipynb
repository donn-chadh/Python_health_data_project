```python
import pandas as pd

df = pd.read_csv('C:/Users/djenb/Desktop/Portfolio/health_data_project/data/raw/heart_rate_tracker.csv', header=1)

# Inspect the data
df.shape
df.columns
df.head()
df.dtypes

df.columns = df.columns.str.replace('com.samsung.shealth.tracker.heart_rate.', '')
df.columns = df.columns.str.replace('com.samsung.health.heart_rate.', '')

# Columns to keep
df = df[['start_time', 'heart_rate', 'time_offset']]

# Convert start_time to datetime format
df['start_time'] = pd.to_datetime(df['start_time'])

# Rename columns for readability
df = df.rename(columns={'start_time': 'recorded_at', 'heart_rate': 'BPM', 'time_offset': 'timezone'})

# Check for missing values
df.isna().sum()

# Check for suspicious BPM readings
len(df[(df['BPM'] >= 200) | (df['BPM'] <= 40)])

df.to_csv('C:/Users/djenb/Desktop/Portfolio/health_data_project/data/cleaned/heart_rate_clean.csv', index=False)

```
