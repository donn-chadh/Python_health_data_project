```python
import pandas as pd

hr = pd.read_csv('../data/cleaned/heart_rate_clean.csv')
steps = pd.read_csv('../data/cleaned/daily_steps_clean.csv')
fitbit = pd.read_csv('../data/cleaned/kaggle_steps.csv')
uci = pd.read_csv('../data/cleaned/uci_heart_rate.csv')

print(hr.columns)
print(steps.columns)
print(fitbit.columns)
print(uci.columns)

# Compare average daily steps and distance
samsung_avg = steps[['steps', 'distance_km']].mean().round(2)
fitbit_avg = fitbit[['fb_total_steps', 'fb_total_distance_km']].mean().round(2)

print(f"Samsung average daily steps: {samsung_avg['steps']}")
print(f"Fitbit average daily steps: {fitbit_avg['fb_total_steps']}")
print(f"Samsung average daily distance (km): {samsung_avg['distance_km']}")
print(f"Fitbit average daily distance (km): {fitbit_avg['fb_total_distance_km']}")

# Compare max heart rate
samsung_max_hr = hr['max_BPM'].mean().round(2)
uci_max_hr = uci['max_heart_rate'].mean().round(2)

print(f"\nSamsung average daily max BPM: {samsung_max_hr}")
print(f"UCI benchmark average max heart rate: {uci_max_hr}")
```
