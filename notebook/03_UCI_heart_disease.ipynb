```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
df = pd.DataFrame(heart_disease.data.features)

# Inspecting data 
print(df['thalach'].describe())
print(df.shape)
print(df.columns)
print(df.head())
print(df.dtypes)
print(df.isna().sum())

# Columns to keep + rename
df = df[['age', 'sex', 'thalach']]
df = df.rename(columns={'thalach': 'max_heart_rate'})

# Create age ranges enabling analysis 
df['age_range'] = pd.cut(df['age'], bins=[0, 20, 30, 40, 50, 60, 70, 100], labels=['<20', '21-30', '31-40', '41-50', '51-60', '61-70', '70+'])

# Filter for my age range
df = df[df['age_range'] == '31-40']

# Filter for male data
df = df[df['sex'] == 1]

# Validate
print(df['max_heart_rate'].describe())

# Saving to CSV file
df.to_csv('../data/cleaned/uci_heart_rate.csv', index=False)
```
